```python
# Replacing 'Select' with NaN (Since it means no option is selected)
df = df.replace('Select',np.nan)
```

## Sanity Check

```python
# Dropping the Columns which has high Null value precentage
leads_df= df.loc[:, (df.isnull().sum()/len(df)*100) < 35]
```

```python
leads_df.head()
```

```python
leads_df.shape
```

```python
# removing the column Prospect ID; Lead Number is a unique value already
leads_df = leads_df.drop('Prospect ID', axis=1)

# removing the five columns which are constant across all rows; they don't add any value to the model
leads_df = leads_df.drop('Magazine', axis=1)
leads_df = leads_df.drop('Receive More Updates About Our Courses', axis=1)
leads_df = leads_df.drop('Update me on Supply Chain Content', axis=1)
leads_df = leads_df.drop('Get updates on DM Content', axis=1)
leads_df = leads_df.drop('I agree to pay the amount through cheque', axis=1)
```

```python
# Generate a summary of the Data frame after dropping the columns.
summary = dfSummary(leads_df)

# Print the summary
summary
```

##  Replacing the Null Values

```python
# Replacing the Missing values in the Categorical Columns with MODE.
# What matters most to you in choosing a course
leads_df['What matters most to you in choosing a course'] = leads_df['What matters most to you in choosing a course'].fillna(leads_df['What matters most to you in choosing a course'].mode()[0])
# Country
leads_df['Country'] = leads_df['Country'].fillna(leads_df['Country'].mode()[0])
# What is your current occupation
leads_df['What is your current occupation'] = leads_df['What is your current occupation'].fillna(leads_df['What is your current occupation'].mode()[0])
# Country
leads_df['Country'] = leads_df['Country'].fillna(leads_df['Country'].mode()[0])
# Last Activity
leads_df['Last Activity'] = leads_df['Last Activity'].fillna(leads_df['Last Activity'].mode()[0])
```

```python
#Checking the Null values Percentage 
round(leads_df.isnull().sum()/len(leads_df)*100,2)
```

```python
# Checking the percent of lose if the null values are removed
round(100*(sum(leads_df.isnull().sum(axis=1) > 1)/leads_df.shape[0]),2)
```

```python
Final_data = leads_df[leads_df.isnull().sum(axis=1) <1]
```

```python
#Checking the Null values Percentage 
round(Final_data.isnull().sum()/len(Final_data)*100,2)
```

```python
# Defining a function to get value Counts of the Categorical Columns.
def value_counts(i):
    print(f"{i} Value Counts")
    print( )
    print(Final_data[i].value_counts())
    print("---------------------------------------")
    print( )
```

```python
# Defining a function to get Bar plot of the Categorical Columns.
def barplot(i):
    fig = px.bar(x=Final_data[i].value_counts().index,y=Final_data[i].value_counts().values,text_auto='.2s')
    fig.update_layout(showlegend=False, title_text=f"{i} Distribution",title_x=0.5)  # add title here
    fig.update_xaxes(title_text=i,showgrid=False)#updating the X_axis1 title
    fig.update_yaxes(title_text="Counts",showgrid=False)#updating the X_axis1 title
    fig.update_traces(marker_color='blue')  # change color to red
    fig.show()
```
